```python
df = pd.read_excel('../data/캔디_생산실적.xlsx')
df.head()
```

```python
df.iloc[:, 0:25].isnull().sum()
```

```python
data = df.head(1)
data
```

```python
data.columns[0:25].tolist()
data.groupby(data.columns[0:25].tolist())
```

```python
data.unstack(level=0)

data.unstack(level=2)
```

```python
tf = data.melt(id_vars= data.columns[0:25], var_name="원재료명_N", value_name="원재료").dropna(subset=['원재료'])
tf
```

```python
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm

for i in tqdm(range(1, len(df))) : #range(len(df)):
    tf = tf.append(df[(i+0):(i+1)].melt(id_vars= data.columns[0:25], var_name="원재료명_N", value_name="원재료").dropna(subset=['원재료']))
```

```python
tf.head(20)
tf.tail(20)
```

```python
tf.to_excel('../data/캔디_생산실적_logntype.xlsx')
```
