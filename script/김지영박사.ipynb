# 첫번째 의뢰작업

```python
df = pd.read_excel('../data/기호식품 원재료명 정리_1차.xlsx')
df
```

```python
df.iloc[:, 0:15].head()
```

```python
data = df.head(1)
data
```

```python
data.iloc[:, 0:15].head()
```

```python
data.columns[0:13]
```

```python
tf = data.melt(id_vars= data.columns[0:13], var_name="순서", value_name="원재료").dropna()
tf
```

```python
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm

for i in tqdm(range(1, len(df))) : #range(len(df)):
    tf = tf.append(df[(i+0):(i+1)].melt(id_vars= data.columns[0:13], var_name="순서", value_name="원재료").dropna())
```

```python
tf.to_excel('../data/기호식품 원재료명 정리_1차_logntype.xlsx')
```
